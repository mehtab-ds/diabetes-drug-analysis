## grouped analysis

#### feature enginerring 
1. make the age group(<40,40-60,60-80,80+) from age columns
2.  make the bmi group(underweight,normal,overweight,obese)) from bmi columns

```python
#1 
df['age_group'] = pd.cut(df['age'],bins=[0,40,60,80,105],labels=['<40','40-60','60-80','80+'])
```

```python
#2 
df['bmi_group']= pd.cut(df['bmi'],bins=[0,18.5,25,30,100],labels=['underweight','normal','overweight','obese'])
```

#### age_group and hba1c_change 
- conclusion
- In every age group (<40, 40–60, 60–80, 80+), Novodra showed higher average HbA1c reductions compared to Auralin.
- The difference was consistent across groups, meaning Novodra’s advantage did not depend on patient age.
- Error bars overlapped slightly, but the overall pattern was stable: Novodra outperformed Auralin in all categories.
- No age group showed reduced effectiveness — both drugs worked across ages, but Novodra was stronger.


```python
sns.barplot(x='age_group', y='hba1c_change', hue='dosage_type', data=df, errorbar='sd')
plt.title("Average HbA1c Change by Age Group and Drug")
```

#### bmi_group and hba1c_change
- conclusion
- In every BMI category (Underweight, Normal, Overweight, Obese), Novodra produced higher average HbA1c reductions compared to Auralin.
- The difference was consistent across groups, showing Novodra’s advantage was not dependent on body size.
- Error bars overlapped slightly, but the overall pattern remained stable: Novodra outperformed Auralin in all BMI ranges.
- HbA1c improvements were not explained by BMI group — the drug itself was the key factor driving effectiveness.


```python
sns.barplot(x='bmi_group',y='hba1c_change',data=df,errorbar='sd',hue='dosage_type')
plt.title("Average HbA1c Change by BMI Group and Drug")
```
